```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
museum_visitors = pd.read_csv('museum-visitors.csv') # reading dataset found on Kaggle
museum_visitors.head(5) # cleaned data in Excel before this
```

```python
plt.plot(museum_visitors.index, museum_visitors.America_Tropical_Interpretive_Center, "r")
plt.plot(museum_visitors.index, museum_visitors.Chinese_American_Museum, "g")
plt.legend(["American Tropical Interpretive Center", "Chinese American Museum"])
plt.xlabel("Month (in months over span of 5 1/2 years)")
plt.ylabel("Amount of Visitors")
plt.title("Fig.1: American vs. Chinese Exhibits Visitors")
plt.show() # red indicates a decrease while green indicates a increase.
```

---
## Percentage of Change Analysis
---

```python
American_Percentages=museum_visitors.America_Tropical_Interpretive_Center / museum_visitors.America_Tropical_Interpretive_Center.iloc[0] * 100
American_Percentages.head(6) # calculating the percentage of change in American visitors from the first month to the last month 
```

```python
Chinese_Percentages = museum_visitors.Chinese_American_Museum / museum_visitors.Chinese_American_Museum.iloc[0] * 100
Chinese_Percentages.head(6) # calculating the percentage of change in Chinese visitors from the first month to the last month
```

```python
plt.plot(museum_visitors.index, American_Percentages, "r")
plt.plot(museum_visitors.index, Chinese_Percentages, "g")
plt.title("Fig.2: Chinese vs. American Exhibits: % Change in Visits")
plt.legend(["American Tropical Interpretive Center", "Chinese American Museum"])
plt.xlabel("Month (in months over span of 5 1/2 years)")
plt.ylabel("Change in Visitors (First Month = 100%)")
plt.show() # red indicates a decrease while green indicates a increase.
```

```python
Chinese_Percentages_Sum = Chinese_Percentages.sum()
Chinese_Average = Chinese_Percentages_Sum / 65
print(Chinese_Average)
```

```python
American_Percentages_Sum = American_Percentages.sum()
American_Average = American_Percentages_Sum / 65
print(American_Average)
```

The Chinese Exhibit shows an increase of **74.37%** over **65** months.   
The American Tropical Interpretive Exhibit show a slight overall increase of **3.74%** over **65** months.   
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; -There was a *significant fluctuation* in visitors.   
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; -A *potential, decreasing trend* can be seen in Figures 1 & 2 above.
